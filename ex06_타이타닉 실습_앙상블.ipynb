```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns 
import warnings
warnings.filterwarnings('ignore')
```

### 앙상블 모델 사용하기
- 머신러닝에서 성능이 좋은 모델

#### RandomForest 모델 사용

```python
# 모델 꺼내오기 
from sklearn.ensemble import RandomForestClassifier

# 모델 생성
rforest_model=RandomForestClassifier()
```

```python
# 교차 검증
from sklearn.model_selection import cross_val_score
result=cross_val_score(rforest_model, X_train, y_train, cv=5)

# 학습 진행 => cv=5 
rforest_model.fit(X_train,y_train)

# 학습한 모델로 예측 
pre=rforest_model.predict(X_test)

# 결과 확인 
result, result.mean(), pre
```

```python
# 평가결과를 kaggle 사이트에서 확인해보기 
submission_rforest=pd.read_csv('./data/gender_submission.csv')
submission_rforest['Survived']=pre
submission_rforest.to_csv('rfroest_ta_02.csv',index=False)
submission_rforest
```

### 하이퍼 파라미터 튜닝
- GridSearchCV() : 최적의 하이퍼파라미터 조합을 찾아주는 함수 
    - 1. 궁금한 하이퍼파라미터 정의 / 설정 
    - 2. param = {} 형태로 하이퍼파라미터 저장 

```python
rforest_model2=RandomForestClassifier(random_state=123) # 기본값 100 
```

```python
# 궁금한 하이퍼파라미터 정의 / 설정 
# max_depth = # 8 ~ 88
# max_features = # 0 ~ 1
# min_samples_leaf = # 2 ~ 5 
# n_estimators = # 100 ~ 3333

# param 변수에 정의 
param={'max_depth' : [8, 18, 28, 38], # 48, 58, 68, 78, 88], # 8 ~ 88 
      'max_features' : [0.1,0.4,0.7], # 0 ~ 1 
      'min_samples_leaf' : [2,3,4], # 2 ~ 5 
      'n_estimators' : [100,1100,2100,3100] # 100 ~ 3333 
      }
```

```python
# GridSearchCV() 함수 꺼내오기
from sklearn.model_selection import GridSearchCV

grid=GridSearchCV(rforest_model2,param,cv=5,verbose=3) # verbose : 학습하는 과정을 확인 
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

Ensemble : 
- 여러개의 DT를 적용한 모델 
- 개별 DT가 과대적합이 쉬운 모델
   --> 다양한 DT를 통해서 과대적합 해결

학습유형에 따라 
- 보팅 : 다른 종류의 모델 --> 예측한 결과(투표/평균)
- 배깅 : 같은 종류의 모델 --> 예측한 결과(투표/평균)
             데이터 샘플링 : 데이터간 중복 허용, 랜덤하게 추출
                                     ==> boostrap 분할방식 
- 부스팅 : 같은 종류의 모델 --> 순차적 학습, 앞 모델의 예측 오차에 대한 가중치를 부여해서 뒷 모델이 학습, 오차를 개선하는 방식
             데이터 샘플링 ==> 앞 모델의 오차에 가중치 부여

RandomForest 다양한 트리를 만드는 방법
1. 데이터를 무작위로 선택
2. 특성을 무작위로 선택
